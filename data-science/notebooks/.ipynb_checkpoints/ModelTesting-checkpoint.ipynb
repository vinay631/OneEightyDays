```python
import pandas as pd
import numpy as np

df = pd.read_csv("/Users/vmandal/Downloads/CAX_CANCER_Updated_TRAIN_data.csv")

print df.shape
```

```python
df = df.loc[:, (df != 0).any(axis=0)]
cols = df.columns
print df.loc[:, cols[1:]].head()
```

```python
from sklearn.cross_validation import train_test_split
df = df.replace([np.inf, -np.inf], np.nan)
df = df.dropna()
df -= df.min()
df /= df.max()
print df.shape
print cols[42]
X, y = df.loc[:, cols[1:]], df.TD50_Rat_mg
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
print df.describe()
```

```python
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR

model = SVR(kernel='poly', C=1e3, degree=2)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
import math

y_pred = model.predict(X_train)
#print y_pred
print math.sqrt(mean_squared_error(y_train, y_pred))
```
